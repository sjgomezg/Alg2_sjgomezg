Ejemplo Bioinformático de busqueda por metodo de fuerza bruta


```python
target='AATGAATTCTCGTAACTACATAGCACAAGTAGA'
inicio = None #indice de inicio de la cadena. O(1)
j = 0 #O(1)
for i in range(0, (len(secuence)-1)): #recorremos 'secuence' en su totalidad. O(n)
  if secuence[i] == target[j]: #evalua si secuence en la posicion i es igual a lo que buscamos en la posicion 0 O(1)
    if j == 0: #evaluamos que aun no tengamos la posicion inicial e la subcadena y lo asignamos O(1)
      inicio=i #O(1)
    if j == len(target) -1: #si la totalidad de la subcadena fue encontrada O(1)
      print(inicio, i) #imprimimos el resultado, el indice de inico y final de la cadena. O(1)
      j = 0 # asignamos 0 para seguir la busqueda O(1)
    j += 1 #se evalua caracter por caracter en busca de igualdad completa O(1)
  else: #O(1)
    j = 0 #aun no se ha encontrado un caracter para iniciar la busqueda O(1)
    #O(1)+O(1)+O(n[O(1)+O(1)+O(1)+O(1)+O(1)+O(1)+O(1)+O(1)+O(1)]) = O(n)
a=secuence.find(target)
print(a)
```
